## 2. Using The Search Engines (Content-Based)
Importing the functions as done below automatically imports the datasets and libraries they depend on. (No need to explicitly load the datasets)

#### I. Search Engine For Book Titles

```python
from search_engines import search_titles

# Simulate a user input for the search query
search_results = search_titles("Harry Potter")

search_results[['book_title', 'book_author', 'year_of_publication', 'publisher', 'image_url_m']]
```

#### II. Search Engine For Book Authors

```python
from search_engines import search_authors

search_results = search_authors("Octavia Butler")

search_results[['book_title', 'book_author', 'year_of_publication', 'publisher', 'image_url_m']]
```

```python
# Searching for a flagged author:
search_results = search_authors("john grisham")
search_results[['book_title', 'book_author', 'content_warning']]
```

#### III. Search Engine Using Multiple Features

```python
from search_engines import search_all

search_results = search_all("george orwell 1984 sci-fi")

search_results[['book_title', 'book_author', 'year_of_publication', 'publisher', 'image_url_m']]
```
