```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
books_ratings = pd.read_csv('data/books_rated.csv', sep=";", encoding="utf-8-sig")
```

```python
# Define a set of words to exclude from the count (e.g., prepositions, articles)
exclude_words = {'a', 'an', 'the', 'of', 'in', 'on', 'at', 'by', 'for', 'to', 'with', 'as', 'and', 'but', 'or', 'nor', 'so', 'yet'}

def split_and_extract_filtered_phrases(text):
    # Split the genre string at commas
    parts = text.split(',')
    
    # Initialize lists to hold results
    split_genres = [part.strip() for part in parts]
    filtered_phrases = []

    # Filter phrases based on word count excluding specific words
    for part in split_genres:
        # Strip whitespace and split into words
        words = part.split()
        # Filter out excluded words
        filtered_words = [word for word in words if word.lower() not in exclude_words]
        
        # Include phrase if it has more than 3 words after filtering
        if len(filtered_words) > 3:
            filtered_phrases.append(part)
    
    return split_genres, filtered_phrases

# Apply to the 'genre' column
books_ratings[['genre', 'annotations']] = books_ratings['genre'].apply(
    lambda text: pd.Series(split_and_extract_filtered_phrases(text))
)

books_ratings.head()
```

```python
# Apply also to the 'genre' column in books_users_ratings
books_users_ratings[['genre', 'annotations']] = books_users_ratings['genre'].apply(
    lambda text: pd.Series(split_and_extract_filtered_phrases(text))
)

books_users_ratings.head()
```

```python
# Apply genre mapping:

# Define the genre mapping
genre_mapping = {
    'fiction': 'Fiction',
    'romance': 'Romance',
    'historical': 'Historical Fiction',
    'thriller': 'Thriller',
    'suspense': 'Thriller',
    'juvenile': 'Juvenile Fiction',
    'children': "Children's Literature",
    'guidebook': 'Guidebooks',
    'self-help': 'Self-Help',
    'poetry': 'Poetry',
    'music': 'Music',
    'unknown': 'Unknown',
    'science': 'Science',
    'Scientific': 'Science',
    'mental health': 'Psychology',
    'psychology': 'Psychology'
}

# Function to return all matching genres for each list of genres
def categorize_genres(genre_list):
    categorized_genres = []
    for genre in genre_list:
        genre_lower = genre.lower()  # Convert to lowercase for case-insensitive matching
        matched_categories = []  # List to hold all matching categories
        for keyword, category in genre_mapping.items():
            if keyword in genre_lower:  # Check if keyword is in genre string
                matched_categories.append(category)  # Add matching category to list
        
        if matched_categories:
            categorized_genres.extend(matched_categories)
        else:
            categorized_genres.append('Other')  # Append 'Other' if no match found

    return list(set(categorized_genres))  # Remove duplicates by converting to set and back to list

# Apply the function to the 'genre' column
books_ratings['genre'] = books_ratings['genre'].apply(categorize_genres)
```

```python
# Apply to books_users_ratings as well
books_users_ratings['genre'] = books_users_ratings['genre'].apply(categorize_genres)
```

```python
# Only keep annotations which include verbs:

import re
from nltk import pos_tag, word_tokenize
from nltk.corpus import stopwords

# Ensure you have the necessary NLTK resources
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')
nltk.download('stopwords')

# Define a set of stopwords
stop_words = set(stopwords.words('english'))

# Function to check if a phrase contains at least one verb
def contains_verb(text):
    tokens = word_tokenize(text.lower())
    tagged_tokens = pos_tag(tokens)
    # Check if any token is a verb
    return any(tag.startswith('VB') for word, tag in tagged_tokens if word not in stop_words)

# Function to filter annotations
def filter_annotations(annotations):
    # Filter out phrases that contain at least one verb
    return [annotation for annotation in annotations if contains_verb(annotation)]

# Apply the function to the 'annotations' column
books_ratings['annotations'] = books_ratings['annotations'].apply(filter_annotations)
```

```python
# Define a function to join a list of strings into a single string separated by periods
def join_with_periods(lst):
    return '.'.join(lst)
```

```python
# Apply the function to the 'annotations' column
books_ratings['annotations'] = books_ratings['annotations'].apply(join_with_periods)
```

```python
# Now do the same for books_users_ratings
books_users_ratings['annotations'] = books_users_ratings['annotations'].apply(join_with_periods)
```
