### Baseline model: our book recommender recommends the books with the highest average ratings & highest rating counts ###

```python
data = pd.read_csv('data/books_rated.csv', sep=";", encoding="utf-8-sig")  
data.head()
```

```python
# sort highest average ratings

high_ratings = data.sort_values(by='avg_rating', ascending=False).head(10)
high_ratings
                                                
```

```python
#sort high rating count

high_count = data.sort_values(by='active_ratings', ascending=False).head(10)
high_count
```

```python
recommendations = data.sort_values(by=['avg_rating', 'active_ratings'], ascending=[False, False])
top_10_recs = recommendations.head(10)
top_10_recs
```
